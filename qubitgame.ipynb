```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

GATES = ["X", "H", "Z"]

def apply_gate(qc, gate):
    if gate == "X":
        qc.x(0)
    elif gate == "H":
        qc.h(0)
    elif gate == "Z":
        qc.z(0)

def get_player_move(player):
    while True:
        move = input(f"{player}, choose a gate (X / H / Z): ").upper()
        if move in GATES:
            return move
        else:
            print("Invalid move. Choose X, H, or Z")

def main():
    qc = QuantumCircuit(1, 1)

    print("\n🎮 Quantum Turn-Based Game Started!")
    print("Allowed gates: X, H, Z")
    print("Total Rounds: 3\n")

    rounds = 3

    for r in range(rounds):
        print(f"--- Round {r+1} ---")

        move1 = get_player_move("Player A")
        apply_gate(qc, move1)

        move2 = get_player_move("Player B")
        apply_gate(qc, move2)

        print()

    qc.measure(0, 0)

    simulator = AerSimulator()
    compiled_circuit = transpile(qc, simulator)
    job = simulator.run(compiled_circuit, shots=1)
    result = job.result()
    counts = result.get_counts()

    print("Final measurement:", counts)

    if '1' in counts:
        print("🔥 Player A wins!")
    else:
        print("❄ Player B wins!")

    print("\nFinal Quantum Circuit:")
    print(qc)

if __name__ == "__main__":
    main()
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

GATES = ["X", "H", "Z"]

def apply_gate(qc, gate):
    if gate == "X":
        qc.x(0)
    elif gate == "H":
        qc.h(0)
    elif gate == "Z":
        qc.z(0)

def get_player_move(player):
    while True:
        move = input(f"{player}, choose a gate (X / H / Z): ").upper()
        if move in GATES:
            return move
        else:
            print("Invalid move. Choose X, H, or Z")

def main():
    qc = QuantumCircuit(1, 1)

    print("\n🎮 Quantum Probability-Based Game Started!")
    print("Allowed gates: X, H, Z")
    print("Total Rounds: 3\n")

    rounds = 3

    for r in range(rounds):
        print(f"--- Round {r+1} ---")

        move1 = get_player_move("Player A")
        apply_gate(qc, move1)

        move2 = get_player_move("Player B")
        apply_gate(qc, move2)

        print()

    # ---- Probability Calculation ----
    state = Statevector.from_instruction(qc)

    p0 = abs(state[0])**2
    p1 = abs(state[1])**2

    print("📊 Outcome Probabilities:")
    print(f"P(|0⟩) = {p0:.4f}")
    print(f"P(|1⟩) = {p1:.4f}\n")

    if p1 > p0:
        print("🔥 Player A WINS (|1⟩ more likely)")
    elif p0 > p1:
        print("❄ Player B WINS (|0⟩ more likely)")
    else:
        print("🤝 DRAW (Equal probabilities)")

    qc.measure(0, 0)

    print("\nFinal Quantum Circuit:\n")
    print(qc)

if __name__ == "__main__":
    main()
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

GATES = ["X", "H", "Z"]
STATES = ["0", "1", "+", "-"]

def prepare_initial_state(qc, state):
    if state == "1":
        qc.x(0)
    elif state == "+":
        qc.h(0)
    elif state == "-":
        qc.x(0)
        qc.h(0)

def apply_gate(qc, gate):
    if gate == "X":
        qc.x(0)
    elif gate == "H":
        qc.h(0)
    elif gate == "Z":
        qc.z(0)

def get_gate(player):
    while True:
        move = input(f"{player}, choose a gate (X / H / Z): ").upper()
        if move in GATES:
            return move
        else:
            print("Invalid gate. Choose X, H, or Z")

def get_initial_state():
    print("Choose initial qubit state:")
    print("0  → |0⟩")
    print("1  → |1⟩")
    print("+  → |+⟩  (superposition)")
    print("-  → |−⟩  (superposition)")

    while True:
        s = input("Your choice (0 / 1 / + / -): ")
        if s in STATES:
            return s
        else:
            print("Invalid choice.")

def main():
    qc = QuantumCircuit(1, 1)

    print("\n🎮 Quantum Turn-Based Strategy Game\n")

    init_state = get_initial_state()
    prepare_initial_state(qc, init_state)

    print(f"\nInitial qubit prepared as |{init_state}⟩\n")

    rounds = 3

    for r in range(rounds):
        print(f"--- Round {r+1} ---")

        a = get_gate("Player A")
        apply_gate(qc, a)

        b = get_gate("Player B")
        apply_gate(qc, b)

        print()

    # Probability calculation
    state = Statevector.from_instruction(qc)
    p0 = abs(state[0])**2
    p1 = abs(state[1])**2

    print("📊 Final Probabilities:")
    print(f"P(|0⟩) = {p0:.4f}")
    print(f"P(|1⟩) = {p1:.4f}\n")

    if abs(p0 - p1) < 1e-6:
        print("🤝 RESULT: DRAW")
    elif p1 > p0:
        print("🔥 RESULT: Player A WINS")
    else:
        print("❄ RESULT: Player B WINS")

    qc.measure(0, 0)

    print("\nFinal Quantum Circuit:\n")
    print(qc)

if __name__ == "__main__":
    main()
```

```python
import random
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

GATES = ["X", "H", "Z"]
STATES = ["0", "1", "+", "-"]

def prepare_initial_state(qc, state):
    if state == "1":
        qc.x(0)
    elif state == "+":
        qc.h(0)
    elif state == "-":
        qc.x(0)
        qc.h(0)

def apply_gate(qc, gate, player=""):
    qc_name = "Player" if player == "Player" else "Computer"
    print(f"{qc_name} applies: {gate}")
    if gate == "X":
        qc.x(0)
    elif gate == "H":
        qc.h(0)
    elif gate == "Z":
        qc.z(0)

def get_player_gate():
    while True:
        g = input("Choose your gate (X / H / Z): ").upper()
        if g in GATES:
            return g
        else:
            print("Invalid gate. Choose X, H, or Z")

def get_initial_state():
    print("Choose initial qubit state:")
    print("0  → |0⟩")
    print("1  → |1⟩")
    print("+  → |+⟩")
    print("-  → |−⟩")

    while True:
        s = input("Your choice (0 / 1 / + / -): ")
        if s in STATES:
            return s
        else:
            print("Invalid choice")

def main():
    qc = QuantumCircuit(1, 1)

    print("\n🎮 Quantum Strategy Game — Player vs Computer 🤖\n")

    init_state = get_initial_state()
    prepare_initial_state(qc, init_state)

    print(f"\nInitial qubit prepared as |{init_state}⟩\n")

    rounds = 3

    for r in range(rounds):
        print(f"--- Round {r+1} ---")

        player_gate = get_player_gate()
        apply_gate(qc, player_gate, "Player")

        comp_gate = random.choice(GATES)
        apply_gate(qc, comp_gate, "Computer")

        print()

    # Probability calculation
    state = Statevector.from_instruction(qc)
    p0 = abs(state[0])**2
    p1 = abs(state[1])**2

    print("📊 Final Probabilities:")
    print(f"P(|0⟩) = {p0:.4f}")
    print(f"P(|1⟩) = {p1:.4f}\n")

    if abs(p0 - p1) < 1e-6:
        print("🤝 RESULT: DRAW")
    elif p1 > p0:
        print("🔥 RESULT: YOU WIN")
    else:
        print("❄ RESULT: COMPUTER WINS")

    qc.measure(0, 0)

    print("\nFinal Quantum Circuit:\n")
    print(qc)

if __name__ == "__main__":
    main()
```
